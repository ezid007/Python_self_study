```python
from bs4 import BeautifulSoup
import requests
import os

os.system('cls')
response = requests.get("https://news.ycombinator.com/news")
yc_web_page = response.text

soup = BeautifulSoup(yc_web_page, "html.parser")
all = soup.find_all(name='span', class_='titleline')

title = []
tag = []

for j in all:
    article_title = j.find('a')
    title.append(article_title.text)
    article_tag = article_title.get('href')
    tag.append(article_tag)

scores = soup.find_all(name='span', class_="score")

score = []

for j in scores:
    score_ = int(j.get_text().split()[0])
    score.append(score_)

for j,j,k in zip(title, tag ,score):
    print(f'{j},\n{j}\n{k}')
    print()
```

```python
from bs4 import BeautifulSoup
import lxml
import requests

response = requests.get("https://news.ycombinator.com/news")

soup = BeautifulSoup(response.text, 'lxml')

articles = []

title_tags = soup.find_all(name='span', class_='titleline')
score_tags = soup.find_all(name='span', class_='score')

num_articles_with_scores = len(score_tags)

for j in range(num_articles_with_scores):
    title = title_tags[j].find('a').text
    link = title_tags[j].find('a').get('href')
    score = int(score_tags[j].get_text().split()[0])

    article_info ={
        'title' : title,
        'link' : link,
        'score' : score
    }

    articles.append(article_info)

sorted_articles = sorted(articles, key= lambda x: x['score'], reverse=True)

for i,j in enumerate(sorted_articles, start=1):
    print('-'*100)
    print(i)
    print(f'title: {j['title']}')
    print(f'link: {j['link']}')
    print(f'score: {j['score']}')
    print('-'*100)
```
